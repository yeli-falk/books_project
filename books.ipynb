# Part 1: Character's names extraction

Take a txt file as the input and output the name of the characters from the book. We are going to use the book '1984' by George Orwell. The desired characters are **Winston Smith, Julia , O'Brien, Big Brother, Mr. Charrington** and **Emmanuel Goldstein**. 

```python
book_path=r'C:\Users\hifin\Desktop\books_project\books\Ninety-eighty-four.txt'
f = open (book_path)
wordlist=dict() #dictionary containing each word and the number of occurences in the text
word_count=0
book = ''
for line in f:
    words=line.split()
    for w in words:
        word=w.lower().strip(',"!)(*-;:?.\/')
        book+=' '+word
        if word in wordlist.keys():
            wordlist[word]+=1
        else:
            wordlist[word]=1
        word_count+=1

print(wordlist)
```

```python
#trivial analysis of the text
print(f'This book has total {word_count} words, with {len(wordlist)} unique words')
values = [i for i in wordlist if wordlist[i]>100]
print(f'This words were used more than 100 times', values)
word_to_search='karl marx'
if word_to_search in wordlist.keys():
    print(f'This book used word {word_to_search} {wordlist[word_to_search]} times')
else:
    print(f'The word {word_to_search} was not found')
```
